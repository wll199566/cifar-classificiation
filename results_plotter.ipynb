```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in the history
with open("./records/resnet_n_3_lr_0.001_wd_0.0_mmt_0.0//history.pkl", "rb") as fin:
    history = pickle.load(fin)
```

```python
history.keys()
```

```python
# plot loss
plt.plot(history["train_loss"], label="train loss")
plt.plot(history["valid_loss"], label="valid loss")
plt.legend()
plt.show()
```

```python
# plot accuracy
plt.plot(history["train_acc"], label="train accuracy")
plt.plot(history["valid_acc"], label="valid accuracy")
plt.legend()
plt.show()
```
